# Equality constrains optimization

### Langrange mutiplier method

### 1.Theaorem:

##### The Langrange is the method for finding the extrema of fucntion of multivariables restricted to given subset.  The method is an optimization problem of form:$$minimize_x f(\vec{r})$$ $$g(\vec{r})=0$$

##### The theorem is : the minum if f is found whent its gradient poin in the sane direction as gradient f.
In other words,when $$\nabla{f(\vec{r})}=\alpha\nabla{g(\vec{r})}$$


  If we have to maximasi function $W=W(x,y)$ by given constraing $\large{G(x,y)=C}$ where C is constants:<br> maximise $\large{W=W(x,y)}$<br> subject to $\large{G(x,y)}=C$ <br>
 the theorem says that exists  the function W miximise G when x,y satisfy ecuation $$\large{\nabla{W}-\gamma.\nabla{G}}=0 $$  $\large{\gamma}$ wich is called langrange mutiplier such that 

6### Proof :
The equation $\large{g(x,y)}=C$ can be expressed as $$\large{x=x(k)}$$   $$\large {y=y(k)} $$ which is justi curve in xy plane.Then we can write  $\large{f=f(x,y)}$<br> subject to $\large{g(x,y)}=C$ <br> as  and we just miximazie $$\large{f=f(x(k),y(k)}$$
then from theorems of Rolles we have $$\frac{df(x(k),y(k)}{dk}=0$$
lets to find the derivate of W $$1) \large{\frac{df(x(k),y(k))}{dk}=\frac{\partial W }{\partial x}.\frac{dx}{dk} + \frac{\partial f}{\partial y }.\frac{dy}{dk}}=0 $$
and $\large{g(x,y)}=C$ $$ 2) \large{\frac{dg(x(k),y(k))}{dk}=\frac{\partial g }{\partial x}.\frac{dx}{dk} + \frac{\partial g}{\partial y }.\frac{dy}{dk}}=0$$
1) and 2) can be expressed as 
$$\large{\begin{pmatrix} 
\frac{\partial f}{\partial x} & \frac{\partial f }{\partial y} \\
\frac{\partial g }{\partial x} & \frac{\partial g }{\partial y}
\end{pmatrix} *
\begin{pmatrix} 
\frac{dx}{dk} \\
\frac{dx}{dy}
\end{pmatrix} =
\begin{pmatrix} 
0 \\
0 \\
\end{pmatrix}
\quad }$$ <br>
<br>
This matrix satisfy  <a href='https://nbviewer.jupyter.org/github/Daodavid93/Machine-Learning/blob/master/Clasification/SVN/THEOREMS%20LINEAR%20DEPENDECY.ipynb'> Theorems 1 </a> 
and we can say the vectors <br>
$\large{\vec{\nabla}=\frac{\partial f}{\partial x}.\vec{e1} + \frac{\partial g}{\partial y }.\vec{e2}}$ and 
$\large{\vec{\nabla g}=\frac{\partial g }{\partial x}.\vec{e1} + \frac{\partial g}{\partial y }.\vec{e2}}$   are linear dependent and exists $\large{\gamma_1}$ and  $\large{\gamma_2}$ such that 
 $$\large{ \gamma_2.\vec{\nabla}.f+ \gamma_2.\vec{\nabla}.g }=0$$ 
 $$\large{\vec{\nabla}.f -  \gamma .\vec{\nabla}.g }=0$$ where $\gamma = -\frac{\gamma_1}{\gamma_2}$

$ \frac{\partial f }{\partial x}=2.x$   ,  $ \frac{\partial f }{\partial y}=2.y$ <br>

# Multiple constaint

Let to have the fuction of 3 variable $\large{f=f(x,y,z}$ and $\large{n=n(x,y,z}$
then the theorems is :
  $$minimize_x f(\vec{r})$$ $$g(\vec{r})=0$$
  $$\nabla{f(\vec{r})}=\alpha\nabla{g(\vec{r})}$$
  $$\large{\nabla{f}-\gamma.\nabla{g} -\beta.\nabla{n}}=0$$ 
  

### proof
let's to have $\large{f=f(x,y,z)}$ and constraints   $\large{g(x,y,z)}$  $\large{n(x,y,z)=0}$
n and g represents surfaces wich ,and exists some curve $\large{x(k)}$ and $\large{y(k)}$ , $\large{z(k)}$ 
for every k $\large{x,y,z \in g and n}$ , for this reason we can rewrite   $\large {f=f(x,y,z)=f=f(x(k),y(k),z(k))}$
then the minimization problem can be reduce to $\large{f'(x(k),y(k),z(k))=0}$
$$f'(\vec{r})=\frac{\partial f}{\partial x}.\frac{dx}{dk}+\frac{\partial f}{\partial y}.\frac{dy}{dk}+\frac{\partial f}{\partial z}.\frac{dz}{dk}=0 $$ 
$$g'(\vec{r})=\frac{\partial g}{\partial x}.\frac{dx}{dk}+\frac{\partial g}{\partial y}.\frac{dy}{dk}+\frac{\partial g}{\partial z}.\frac{dz}{dk}=0 $$ 
$$n'(\vec{r})=\frac{\partial n}{\partial x}.\frac{dx}{dk}+\frac{\partial n}{\partial y}.\frac{dy}{dk}+\frac{\partial n}{\partial z}.\frac{dz}{dk}=0 $$
##### in matrix form 
$$\large{\begin{pmatrix} 
\frac{\partial f}{\partial x} & \frac{\partial f }{\partial y} &  \frac{\partial f }{\partial z} \\
\frac{\partial g }{\partial x} & \frac{\partial g }{\partial y} & \frac{\partial g }{\partial z} \\
\frac{\partial n }{\partial x} & \frac{\partial n }{\partial y} & \frac{\partial n }{\partial z} 
\end{pmatrix} .
\begin{pmatrix} 
\frac{dx}{dk} \\
\frac{dx}{dy} \\
\frac{dx}{dz}
\end{pmatrix} =
\begin{pmatrix} 
0 \\
0 \\
0
\end{pmatrix}
\quad }$$
This matrix A satisfy  <a href='https://nbviewer.jupyter.org/github/Daodavid93/Machine-Learning/blob/master/Clasification/SVN/THEOREMS%20LINEAR%20DEPENDECY.ipynb'> Theorems 1 </a> and from therorems we gets 
that exisist number $\large{\gamma_1}$ and  $\large{\gamma_2}$ such that 
 $$\large{ \gamma_2.\vec{\nabla}.f(\vec r)+ \gamma_2.\vec{\nabla}.g(\vec r) + \gamma_3.\vec{\nabla}.n(\vec r)}=0$$ <br>
 $$\large{\vec{\nabla}.f(\vec r) -  \gamma .\vec{\nabla}.g(\vec r) - \beta{\vec{\nabla}.n(\vec r)}}=0$$ where $\gamma = -\frac{\gamma_2}{\gamma_1}$ $\beta = -\frac{\gamma_3}{\gamma_1}$


### Karush-Kuhn-Tucker Conditions

```python
class Dplot3D:
    """
    
    """
    def __init__(self, _range =20, fig_size=(20,20),elev=10,azim=10,mesh_const=0.1,**kwargs):
        self.fig_size = fig_size
        self.fig = plt.figure()
        self.fig = plt.figure(figsize=fig_size)
        self.ax = self.fig.gca(projection='3d')
       
        #self.ax = fig.add_subplot(111, projection='3d')
        #self.ax.set_aspect("equal")
        self.ax.set_xlabel('X',fontsize=30)
        #self.ax.view_init(elev=elev, azim=azim)
        self.ax.set_ylabel('Y',fontsize=30)
        self.ax.set_zlabel('Z',fontsize=30)
        self.range=_range
        self.mesh = mesh_const
        #plt.xlabel('$\textbf{time} (s)$')
        
    def init_cordinates(self,xlimt=10,ylimt=10,zlimt=10):
        """
        
        """
        #self.ax.quiver(0, 0, 0, 0, 1, 0,arrow_length_ratio=50)
        #self.ax.quiver(0, 0, 0, 1, 0, 0,arrow_length_ratio=50)
        #self.ax.quiver(0, 0, 0, 0, 0,1,arrow_length_ratio=50)
        self.ax.set_xlim([-xlimt, xlimt])
        self.ax.set_ylim([-ylimt, ylimt])
        self.ax.set_zlim([-zlimt, zlimt])
       
        
        
        
    def _calculate_funct(self,func):
        x, y = np.mgrid[-self.range:self.range:self.mesh,-self.range:self.range:self.mesh]
        return  x,y,func(x,y)

        
    def append_funct(self,function,color='b',label=''):
        x,y,z = self._calculate_funct(function)
        self.ax.plot_wireframe(x, y, z,color=color,label=label,rstride=15, cstride=5)
   
    def append_curve(self,x_funct, y_funct, z_funct,_range=5,color='y',label='',linewidth=1.0):
        """
        x= 2+k*2
        y = 3+k*4
        z = 5 + k*4
        """
        
        k = np.linspace(-_range,_range,100)
        x = x_funct(k)
        y = y_funct(k)
        z = z_funct(k)
        if type(z)==int:
            z=np.ones(len(x))*z
        self.ax.plot(x, y, z,color=color,label=label,linewidth=linewidth)    
            
    def append_circle_xy(self,r,z=0,color='b'):
        theta = np.linspace(0, 2 * np.pi, 100)
        x = r * np.sin(theta)
        y = r * np.cos(theta)
        z = np.ones(len(x))*z
        self.ax.plot(x, y, z, color = color)
       
    def plot_plane(self,z):
        pass
    
    def append_v(self,x_0,y_0,x,y,index='1',show_cord=True,font_size=15):
        self.ax.arrow(x_0, y_0, x, y, head_width=0.1, head_length=0.1, fc='black', ec='black')
    
        if show_cord:
        #self.ax.text(x, y-0.2, r'$\vec{r}_{%s}(%1.1f:%1.1f)$' % (index,x, y),fontsize=font_size,color='blue')
           self.ax.text(x, y-0.2, r'$\vec{r}_{%s}(%1.f:%1.f)$' % (index,x, y),fontsize=font_size,color='blue')
        
    def show(self):
        plt.legend()
        plt.show()
   
```
