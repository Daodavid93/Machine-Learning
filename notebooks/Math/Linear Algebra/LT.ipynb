```python
def v_plot(X_1,X_2,plt_range=[-5,5],label_coord=False,pattern_label=r'$\vec{r}_{%s}(%1.f:%1.f)$',**kwargs):
    '''
    X,Y : nd.array,int
    
    
    '''
    #fig, ax = plt.subplots()
    plt.grid()
    ax = plt.gca()
    
    ax.set_aspect('equal')
#     plt.xlim(-np.max(X_1)+2,np.max(X_1)+2)
#     plt.ylim(-np.max(X_2)-2,np.max(X_2)-2)
    plt.xlim(plt_range[0],plt_range[1])
    plt.ylim(plt_range[0],plt_range[1])
    if isinstance(X_1, (list,tuple, np.ndarray)):
        x_0=y_0 = np.zeros(len(X_1))
    else :
        x_0=y_0=0

      #print(len(x_1))
    q = ax.quiver(x_0, y_0, X_1, X_2,**kwargs)
    if label_coord:
            #self.ax.text(x, y-0.2, r'$\vec{r}_{%s}(%1.1f:%1.1f)$' % (index,x, y),fontsize=font_size,color='blue')
        if isinstance(X_1, (list,tuple, np.ndarray)):    
            for index in range(len(X_1)):  
                ax.text(X_1[index],X_2[index]-0.2,pattern_label  % (index,X_1[index], X_2[index]),fontsize=10,color='blue')
        else :
            ax.text(X_1,X_2-0.2,pattern_label  % (0,X_1, X_2),fontsize=10,color='blue')
    return ax        
```

Liniear tranformation <br>


DEFINITION (Linear Transformation): T transformation (or mapping) $A$ from a vector space $V1$ to a vector space $V2$, $T : V1 \rightarrow V2$ is a linear transformation (or a linear operator, a linear map, etc.), if:

* $T(\vec u +\vec v) = T\vec u + T\vec v$ for all vectors $u$,$v \in V1$.

* $T(c\vec u) = cT\vec u$ for all vectors $\vec u \in V1$ and all scalars $c$.

if we want to rotate $x,y$ an angle $\theta$ we will achive the cordinates $x',y'$, <br>

$x =r.cos\phi$    $y =r.sin\phi$

from above graphic we can write that  <br>
<br> <br>

$x' = cos(\theta + \phi )r = cos(\theta)cos(\phi)r - sin(\theta)sin(\phi)r = cos(\theta).x -sin(\theta)y $ <br>
$y' = sin(\theta + \phi )r=cos(\theta)sin(\phi)r + cos(\phi)sin(\theta)r =-sin(\theta).x + cos(\theta)y  $ <br>
in matrix form above eq. look like this <br> 

$\begin{bmatrix}  x' \\ y' \\ \end{bmatrix} = \begin{bmatrix}  cos(\theta) & -sin(\theta)\\ sin(\theta) & cos(\theta) \\ \end{bmatrix} .\begin{bmatrix}  x \\ y \\ \end{bmatrix} $

example if we have vector $\vec r =  \begin{bmatrix}  x' \\ y' \\ \end{bmatrix} $,and we want to ratate it to $45*$ 
then the rotation matrix will looks like this  $ \begin{bmatrix}  cos45^* & -sin45^*\\ sin45^* & cos45^* \\ \end{bmatrix} $ 


```python
R_45 = np.array([[np.cos((45/180)*np.pi), -np.sin((45/360)*2*np.pi)],
                 [np.sin((45/360)*2*np.pi),np.cos((45/360)*2*np.pi)]]) 
```

```python
def rot_matrix(alpha):
    return np.array([[np.cos((alpha/180)*np.pi), -np.sin((alpha/360)*2*np.pi)],
                 [np.sin((alpha/360)*2*np.pi),np.cos((alpha/360)*2*np.pi)]]) 
    
```

```python
R_30 = rot_matrix(30)
v = VectorSpace2D(plt_range=[-30,30])
T = np.array([[2,0],[0,2]]).dot(R_30)
v.transform(T)
```

```python
R_30 = rot_matrix(45)
v = VectorSpace2D(plt_range=[-30,30])
T = np.array([[1,0],[0,1]]).dot(R_30)
v.transform(T)
```
