```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
from sklearn.decomposition import PCA
```

## Covariance : firsts step to idea ,how the different variables can be related

```python
x = np.random.uniform(0,10,100)
y = np.random.uniform(0,10,100)
```

```python
np.cov(x,y)
```

```python
plt.scatter(x,y)
```

```python
f = lambda x,y : 0.4*x+3*y
z = f(x,y)
cov = np.cov([x,y,f(x,y)])
cov
```

```python
fig = plt.figure(figsize=[20,20])
ax = fig.add_subplot(111, projection='3d')
lines = ax.scatter(x, y, f(x,y),\
                   alpha=0.5,edgecolors='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
pca = PCA()
data = np.array([x,y,z]).T
pca.fit(data)
```

```python
explained_variance = pca.explained_variance_
explained_variance
```

```python
tranformed = pca.transform(data)
```

```python
fig = plt.figure(figsize=[20,20])
ax = fig.add_subplot(111, projection='3d')
lines = ax.scatter(tranformed[:,0], tranformed[:,1], tranformed[:,2],\
                   alpha=0.5,edgecolors='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
