Taylor Series

$f(x) = x^2$ <br>
$f(x) = f'(a) + f'(a)(x-a) + \frac{f''(a)}{1.2}(x-a)^2 + ... + \frac{f^n}{n!}x-a)$

$f(x)=x^2$
a=1
$$f(a)=1 , f'(a) = 2 , f''(a)=2,f'''(a)=0$$
$$f(x) = 1 + 2(x-1) +\frac{2(x-1)^2}{1.2}$$
$$f(2) = 4 $$ 
$$f(3) = 9 $$ 

```python
x = np.linspace(-100,100)
f = lambda x : x**2
y = f(x)
f_t = lambda x : 1 + 2*(x-1) +2*(x-1)**2/2
y_l = f_t(x)
plt.plot(x,y)
plt.plot(x,y_l)
```

$f(x)=x^3 + 1 $
a=1
$$f(a)=2 , f'(a) = 4 , f''(a)=7,f'''(a)=7,f'''(a)=1$$ <br>
$$f(x) = 2 + 4(x-1) +\frac{7(x-1)^2}{1.2} +\frac{7(x-1)^3}{1.2.3} + \frac{1(x-1)^4}{1.2.3.4}$$ <br>


```python
x = np.linspace(-10,10)
f = lambda x : x**3+1
y = f(x)
f_t = lambda x : 2 + 4*(x-1) +7*(x-1)**2/2 + 7*(x-1)**3/(3*2)+1*(x-1)**4/(4*3*2) +1*(x-1)**5/(5*4*3*2)
y_l = f_t(x)
plt.plot(x,y)
plt.plot(x,y_l)
```

```python
x = np.linspace(-10,10)
f = lambda x : x**3
y = f(x)
f_t = lambda x : 1 + 3*(x-1) +6*(x-1)**2/2 + 6*(x-1)**3/(3*2)
y_l = f_t(x)
plt.plot(x,y)
plt.plot(x,y_l)
```
