### 1) Tailor Series of function  with one variable

Consider the function $f(x)$ difined by a power series in form :
$$ f(x) = \sum_i^{\infty}c_n(x-a)^n$$
$f(x)$ as expansion $$f(x)= c_0 + c_1.(x-a)^1+ c_2.(x-a)^2+ c_3.(x-a)^3+ c_4.(x-a)^4...$$
we observed that   $f(a)=c_0$ Moreover <br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ $f^{1}(a)=c_1 $ <br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ $f^{2}(a)=2.c_1$ <br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ $f^{3}(a)=3.2.c_1$ <br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ $f^{4}(a)=4.3.2.c_4$ <br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ $f^{5}(a)=5.4.3.2.c_1$ <br>
in general :
$$ f^{n}(a)=n!.c_n$$ or
$$c_n = \frac{f^{n}(a)}{n!} $$ <br> then we achive for taylor function <br>
$$T(x) = f(a) +  \frac{f^{1}(a).(x-a)^1}{1}+\frac{f^{2}(a).(x-a)^2}{2.1}+\frac{f^{1}(a).(x-a)^3}{3.2.1}...\frac{f^{n}(a).(x-a)^n}{n!}$$ <br> <br>
That is called <big>** Tailor series**</big>
if $a=0$ the expanion $ f(x) = \sum_i^{\infty}c_n(x-0)^n= \sum_i^{\infty}c_n(x)^n$   is  called  <big>** Macloren series</big> : <br> <br>
$$f(x)= f^(0)+ \frac{f^{1}(0)}{1}.x^3+\frac{f^{2}(0)}{2.1}.x^2+ \frac{f^{3}(0)}{3.2.1}.x^3...\frac{f^{n}(0)}{n!}.x^n$$ <br> <br>
$$f(x) =\sum_i^{\infty}\frac{f^{n}(a)}{n!}.x^n $$

### Taylor approximation  of function with 2  variables
Let f be an infinitely differentiable function in some open neighborhood around (x, y) = (a, b) <br>
$$f(x_1,x_2)\approx f(a,b)(x-a) + f_{x_1}(a,b)(x-a)+ f_{x_2}(a,b)(x-b) + \frac{1}{2!}[f_{x_1x_1}(a,b)(x-a)^2 +2f_{x_1x_2}(a,b)(x-a)(x-b)+f_{a,b}+f_{x_2x_2}(a,b)(x-b)^2]...$$
or in more compact form :
$$ f(x) = f(a) + [(x-a)\nabla f(a)] +[(x-a)H(x)(x-a)]...$$ 
where $a = <a,b> $ $x=<x_1,x_2>$ and
$$ Hesian \; H = \begin{bmatrix} f_{x_1x_1}&f_{x_1x_2} \\
f_{x_2x_2} & f_{x_2x_1}   \end{bmatrix}=\begin{bmatrix} \frac{\partial^2{f}}{\partial{x_1}\partial{x_1}}  & \frac{\partial^2{f}}{\partial{x_1}\partial{x_2}}  \\ \frac{\partial^2{f}}{\partial{x_2}\partial{x_2}} & \frac{\partial^2{f}}{\partial{x_2}\partial{x_1}}  \end{bmatrix}$$
