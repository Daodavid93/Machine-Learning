# Sigmoid function

### Sigmoid function having a 'S'-shaped curve ot sigmoid curve.It look like this :
$$\sigma(z) = \frac{1}{1+e^{-z}} $$ where $z$ can be euaqual of $$ z = w^T.x + b$$ where $$w=w_1+w_2+..w_n$$ and 
$$ x = x_1 + x_2 +...+x_n$$ and $$b$$ is constatant vector $$w,x,w_0 \in R^n $$ linear space z is euation of hyperplane.
#### Let to investigate $$\sigma= \sigma(w^t.x + b)$$ when $$w = w_1.e_1 + w_2.e2 $$ <br> $$b=b1.e_1+b_2.e_2 $$ $e_1,e_2$ is baces vector

```python
def sigmoid(w,b,range=10):
    """
    
    """
        
    f = lambda x : 1/(1+np.exp(-x))
    x = np.linspace(-range,range)
    x = np.array([x,x]).T
    z = np.array( [ np.array(w).dot(i)+b for i in x])
    result = np.array([f(i) for i in z])
    return z,result
    
   
```

```python
x,y = sigmoid([1,2],0.4,range=30)
plt.plot(x,y)
plt.show()
x,y = sigmoid([0.03,0.005],0.008,range=230)
plt.plot(x,y)
plt.show()
x,y = sigmoid([0.03,0.001],0.008,range=230)
plt.plot(x,y)
plt.show()
x,y = sigmoid([0.03,0.001],8,range=230)
plt.plot(x,y)

x,y = sigmoid([0.003,0.001],8,range=230)
plt.plot(x,y)
```
